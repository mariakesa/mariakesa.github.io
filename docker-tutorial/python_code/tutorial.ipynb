```python
import os
import threading
import time

def worker(name):
    print(
        f"[thread {name}] "
        f"PID={os.getpid()} "
        f"TID={threading.get_native_id()}"
    )
    time.sleep(1)

print(
    f"[main thread] "
    f"PID={os.getpid()} "
    f"TID={threading.get_native_id()}"
)

threads = []
for i in range(3):
    t = threading.Thread(target=worker, args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import threading
import time

counter = 0

def worker():
    global counter
    for _ in range(100_000):
        counter += 1

threads = []
for _ in range(4):
    t = threading.Thread(target=worker)
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Final counter:", counter)
```
