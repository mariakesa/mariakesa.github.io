```python
import os
import threading
import time

class Neuron:
    def __init__(self, weights, inputs, lock, counter):
        self.weights=weights
        self.lock=lock
        self.inputs=inputs
        self.accumulator=0
        self.counter=0

    def work(self, index):
        with self.lock:
            self.counter+=1
            w=self.weights[index]
            input=self.inputs[index]
            mul=w*input
            self.accumulator+=mul
        self.wait()
        return mul
    
    def wait(self):
        if self.counter==len(self.inputs):
            return
        else:
            time.sleep(1)

lock = threading.Lock()
neuron=Neuron([1,2,3],[4,5,6],lock)
print(neuron.lock)
threads=[]

for i in range(3):
    t = threading.Thread(target=neuron.work, args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

neuron.accumulator
```
